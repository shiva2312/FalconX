```python
# Package for scientific computing.
import numpy as np 

# Package for tabular data manipulation and analysis.
import pandas as pd

from gql import gql, Client
from gql.transport.aiohttp import AIOHTTPTransport

from datetime import datetime

from sklearn.preprocessing import normalize

# Plotting libraries.
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## HERE WE WILL GET DATA ##
```

```python
# Select your transport with a defined url endpoint
transport = AIOHTTPTransport(url="https://api.thegraph.com/subgraphs/name/uniswap/uniswap-v2")

# Create a GraphQL client using the defined transport
client = Client(transport=transport, fetch_schema_from_transport=True)
```

```python
# GraphQL query
query = gql(
    """
    query transaction($skip: Int!){
                swaps(where:{pair: "0xd3d2e2692501a5c9ca623199d38826e513033a17"}, 
                           orderBy: timestamp, 
                           orderDirection: desc,
                           first:1000, skip:$skip) {
                    id
                    timestamp
                    amount0In
                    amount1In
                    amount0Out
                    amount1Out
                    pair {
                        token0Price
                        token1Price
                        reserveUSD
                        volumeUSD
                    }
                }
                
                mints(where:{pair: "0xd3d2e2692501a5c9ca623199d38826e513033a17"}, 
                           orderBy: timestamp, 
                           orderDirection: desc,
                           first:1000, skip:$skip) {
                    id
                    timestamp
                    amount0
                    amount1
                    liquidity
                    amountUSD
                }
                
                burns(where:{pair: "0xd3d2e2692501a5c9ca623199d38826e513033a17"}, 
                           orderBy: timestamp, 
                           orderDirection: desc,
                           first:1000, skip:$skip) {
                    id
                    timestamp
                    amount0
                    amount1
                    liquidity
                    amountUSD
                }
            }
"""
)
```

```python
# Dictionary of variables that are passed to the query.
variables = {'skip': 0}

# Swaps.
# range(0, x), x=[1,6]
data_swaps = []
data_mints = []
data_burns = []
for i in range(0, 2):
    curr = await client.execute_async(query, variables)
    
    data_swaps += curr["swaps"]
    data_mints += curr["mints"]
    data_burns += curr["burns"]
    
    variables["skip"] += 1000
```

```python
## Here we will PARSE THE DATA##
```

```python
# Create a dictionary.
columns = ["Timestamp",
           "amount0In",
           "amount0Out",
           "amount1In",
           "amount1Out",
           "Pool_ID",
           "Liquidity",
           "Volume_USD",
           "token0Price",
           "token1Price",]

data = {}

for col in columns:
    data[col] = []

data
```

```python
# Parse the retrieved data.
for entry in data_swaps:
    data[columns[0]] += [datetime.fromtimestamp(int(entry["timestamp"]))]
    data[columns[1]] += [float(entry["amount0In"])]
    data[columns[2]] += [float(entry["amount0Out"])]
    data[columns[3]] += [float(entry["amount1In"])]
    data[columns[4]] += [float(entry["amount1Out"])]
    data[columns[5]] += [entry["id"]]
    data[columns[6]] += [float(entry["pair"]["reserveUSD"])]
    data[columns[7]] += [float(entry["pair"]["volumeUSD"])]
    data[columns[8]] += [float(entry["pair"]["token0Price"])]
    data[columns[9]] += [float(entry["pair"]["token1Price"])]
```

```python
# Convert the dictionary to pandas DataFrame object.
dataset_swaps = pd.DataFrame.from_dict(data)

dataset_swaps
```

```python
dataset_swaps.info()
```

```python
# Buy base currency and sell quote currency, side = 1
# Sell base currency and buy quote currency, side = -1
dataset_swaps["Side"] = np.where(dataset_swaps["amount0Out"] > 0.0, 1, -1)

dataset_swaps["Quantity_Base_Currency"] = np.where(dataset_swaps["amount0Out"] > 0.0, 
                                                   dataset_swaps["amount0Out"], 
                                                   dataset_swaps["amount0In"])

dataset_swaps["Quantity_Quote_Currency"] = np.where(dataset_swaps["amount0Out"] > 0.0, 
                                                    dataset_swaps["amount1In"], 
                                                    dataset_swaps["amount1Out"])

dataset_swaps
```

```python
# Removing extra columns.
dataset_swaps.drop(labels=["amount0In", "amount0Out", "amount1In", "amount1Out"], axis=1, inplace=True)

dataset_swaps
```

```python
## Now we will extract the Liquidity & Volume ##
```

```python
# Create a dictionary.
columns = ["Timestamp",
           "Pool_ID",
           "Base_Currency",
           "Quote_Currency",
           "Volume_USD",
           "Liquidity",
           "Position"]

data = {}

for col in columns:
    data[col] = []

data
```

```python
# Parse the retrieved data.
for entry in data_mints:
    data[columns[0]] += [datetime.fromtimestamp(int(entry["timestamp"]))]
    data[columns[1]] += [entry["id"]]
    data[columns[2]] += [float(entry["amount0"])]
    data[columns[3]] += [float(entry["amount1"])]
    data[columns[4]] += [float(entry["liquidity"])]
    data[columns[5]] += [float(entry["amountUSD"])]
    data["Position"] += [1]
    
for entry in data_burns:
    data[columns[0]] += [datetime.fromtimestamp(int(entry["timestamp"]))]
    data[columns[1]] += [entry["id"]]
    data[columns[2]] += [float(entry["amount0"])]
    data[columns[3]] += [float(entry["amount1"])]
    data[columns[4]] += [float(entry["liquidity"])]
    data[columns[5]] += [float(entry["amountUSD"])]
    data["Position"] += [-1]
```

```python
# Convert the dictionary to pandas DataFrame object.
dataset_liq = pd.DataFrame.from_dict(data)

dataset_liq
```

```python
dataset_liq.sort_values(by="Timestamp", inplace=True)
```

```python
dataset_liq.reset_index(inplace=True)
```

```python
# Liquidity, +ve means added and vice versa
dataset_liq["Liquidity_"] = dataset_liq["Liquidity"] * dataset_liq["Position"]

dataset_liq
```

```python
dataset_liq["Cumulative_Liquidity"] = np.cumsum(dataset_liq["Liquidity_"])

dataset_liq
```

```python
plt.figure(figsize=(10, 5))
plt.hist(np.log(dataset_liq["Liquidity"]), bins=100)
plt.xlabel("Liquidity (log scale)")
plt.ylabel("Frequency")
plt.title("Liquidity Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(dataset_liq["Timestamp"], dataset_liq["Cumulative_Liquidity"])
plt.xlabel("Date")
plt.ylabel("Cumulative Liquidity in US$")
plt.show()
```

```python
## In the above graph liquidity are becoming negative because lots of people have burned their pairs ##
```

```python
# GraphQL query for price of token 0.
query0 = gql(
    """
{
    tokenDayDatas(orderBy: date, 
                  orderDirection: desc,
                  where: {
                        token: "0x1f9840a85d5af5bf1d1762f925bdaddc4201f984"
                    }
    ) {
        date
        priceUSD
    }
}
"""
)

# GraphQL query for price of token 1.
query1 = gql(
    """
{
    tokenDayDatas(orderBy: date, 
                  orderDirection: desc,
                  where: {
                        token: "0xc02aaa39b223fe8d0a0e5c4f27ead9083c756cc2"
                    }
    ) {
        date
        priceUSD
    }
}
"""
)

# Run the queries.
price0 = await client.execute_async(query0)
price1 = await client.execute_async(query1)

# Check the received data.
price0["tokenDayDatas"][0:2]
```

```python
# Parse the recieved data.
data_price = {"Timestamp" : [], "Price_UNI" : [], "Price_WETH" : []}
for entry0, entry1 in zip(price0["tokenDayDatas"], price1["tokenDayDatas"]):
    data_price["Timestamp"] += [datetime.fromtimestamp(int(entry0["date"]))]
    data_price["Price_UNI"] += [float(entry0["priceUSD"])]
    data_price["Price_WETH"] += [float(entry1["priceUSD"])]

# Convert the dictionary to pandas DataFrame object.
dataset_price = pd.DataFrame.from_dict(data_price)

dataset_price
```

```python
# Price of token 0 over time
plt.figure(figsize=(15, 5))
plt.plot(dataset_price["Timestamp"], dataset_price["Price_UNI"])
plt.xlabel("Date")
plt.ylabel("Price in US$")
plt.title("UNI")
plt.show()
```

```python
# Price of token 0 over time.
plt.figure(figsize=(15, 5))
plt.plot(dataset_price["Timestamp"], dataset_price["Price_WETH"])
plt.xlabel("Date")
plt.ylabel("Price in US$")
plt.title("WETH")
plt.show()
```

```python
# Data over last 7 days.
price = np.vstack((dataset_price["Price_UNI"][:7].to_numpy(), 
                   dataset_price["Price_WETH"][:7].to_numpy()))

# Normalize the data for better visualization.
price = normalize(price, axis = 1)

# Price of token 0 and token 1 over time (normalized).
plt.figure(figsize=(10, 5))
plt.plot(dataset_price["Timestamp"][:7], price[0])
plt.plot(dataset_price["Timestamp"][:7], price[1])
plt.xlabel("Date")
plt.ylabel("Price (Normalized)")
plt.legend(["Price UNI", "Price WETH"])
plt.show()
```
